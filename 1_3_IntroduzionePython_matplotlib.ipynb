# Matplotlib

Una parte fondamentale del calcolo scientifico è tracciare i risultati o i dati. Lo strumento in Python più adatto a questo compito è il pacchetto matplotlib.

## Creare un plot

Tracciamo la funzione seno sul dominio x ∈ [0, 2π], colorariamola di verde, inseriamo una griglia e forniamo una legenda significativa e le etichette degli assi. Per fare ciò dobbiamo:
* Importare Numpy in modo da poter sfruttare alcune buone routine numeriche.
* Importare il modulo pyplot di matplotlib. Il modo standard per inserirlo è con il nickname plt (proprio come con numpy quando lo importiamo come np).


```python
import numpy as np
import matplotlib.pyplot as plt

# creiamo un array numpy x (usando il comando np.linspace) e un array numpy y per la funzione seno.
x = np.linspace(0,2*np.pi, 100)
y = np.sin(x)

plt.plot(x,y, 'green', label='The Sine Function') #costruiamo il grafico
plt.grid() #aggiungiamo la griglia
plt.legend() #aggiungiamo la legenda
plt.xlabel("x axis") #aggiungiamo le etichette agli assi x e y
plt.ylabel("y axis")
plt.show() #mostriamo il grafico
```

### Esempio

In questo esempio proveremo a tracciare il grafico delle seguenti curve:
* $y_0 = sin(2πx)$  
* $y1 = cos(2πx) $
* $y_2 = y_0 + y_1 $  
* $y_3 = y_0 − y_1$

nel dominio di x ∈ [0, 1] utilizzando 100 punti equidistanti.

```python
import numpy as np
import matplotlib.pyplot as plt

# %matplotlib inline # you may need this in Jupyter Notebooks
# build the x and y values
x = np.linspace(0,1,100)
y0 = np.sin(2*np.pi*x)
y1 = np.cos(2*np.pi*x)
y2 = y0 + y1
y3 = y0 - y1

# plot each of the functions (notice that they will be on the same axes)
plt.plot(x, y0, 'b-.', label=r"$y_0 = \sin(2\pi x)$")
plt.plot(x, y1, 'r--', label=r"$y_1 = \cos(2\pi x)$")
plt.plot(x, y2, 'g:', label=r"$y_2 = y_0 + y_1$")
plt.plot(x, y3, 'k-', label=r"$y_3 = y_0 - y_1$")

# put in a grid, legend, title, and axis labels
plt.grid()
plt.legend()
plt.title("Grafico di 4 curve")
plt.xlabel('etichetta per asse x')
plt.ylabel('etichetta per asse y')
plt.show()
```

### Esercizio

Tracciare il grafico delle seguenti funzioni:
* $f(x) = x_2$,
* $g(x) = x_3$,
* $h(x) = x_4$.

sullo stesso asse x, nel dominio di x ∈ [0, 1] and y ∈ [0, 1]. Visualizzare la griglia, la legenda, il titolo e le etichette degli assi.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the domain
x = np.linspace(0, 1, 100)

# Define the functions
f_x = x ** 2
g_x = x ** 3
h_x = x ** 4

# Plot the functions
plt.plot(x, f_x, label='f(x) = x^2')
plt.plot(x, g_x, label='g(x) = x^3')
plt.plot(x, h_x, label='h(x) = x^4')

# Add grid, legend, title, and labels
plt.grid(True)
plt.legend()
plt.title('Plot of Functions')
plt.xlabel('x')
plt.ylabel('y')

# Set the limits for both x and y axes
plt.xlim(0, 1)
plt.ylim(0, 1)

# Show the plot
plt.show()
```

## Subplots

Spesso è molto comodo posizionare diversi grafici uno accanto all'altro. Il comando subplots ci consente questa funzionalità. L'idea principale è quella di creare una matrice di grafici vuoti e quindi popolare gli assi con i grafici che vogliamo.

```python
import numpy as np
import matplotlib.pyplot as plt

# same functions as before
x = np.linspace(0,1,100)
y0 = np.sin(2*np.pi*x)
y1 = np.cos(2*np.pi*x)
y2 = y0 + y1
y3 = y0 - y1

# set up the blank matrix of plots
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

# Build the first plot
axes[0,0].plot(x, y0, 'b-.')
axes[0,0].grid()
axes[0,0].set_title(r"$y_0 = \sin(2\pi x)$")
axes[0,0].set_ylim(-2,2)
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")

# Build the second plot
axes[0,1].plot(x, y1, 'r--')
axes[0,1].grid()
axes[0,1].set_title(r"$y_1 = \cos(2\pi x)$")
axes[0,1].set_ylim(-2,2)
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")

# Build the third plot
axes[1,0].plot(x, y2, 'g:')
axes[1,0].grid()
axes[1,0].set_title(r"$y_2 = y_0 + y_1$")
axes[1,0].set_ylim(-2,2)
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("y")

# Build the last plot
axes[1,1].plot(x, y3, 'k-')
axes[1,1].grid()
axes[1,1].set_title(r"$y_3 = y_0 - y_1$")
axes[1,1].set_ylim(-2,2)
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("y")

fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# same functions as before
x = np.linspace(0,1,100)
y0 = np.sin(2*np.pi*x)
y1 = np.cos(2*np.pi*x)
y2 = y0 + y1
y3 = y0 - y1

# set up the blank matrix of plots
fig, ((ax0, ax1),(ax2, ax3)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

# Build the first plot
ax0.plot(x, y0, 'b-.')
ax0.grid()
ax0.set_title(r"$y_0 = \sin(2\pi x)$")
ax0.set_ylim(-2,2)
ax0.set_xlabel("x")
ax0.set_ylabel("y")

# # Build the second plot
ax1.plot(x, y1, 'r--')
ax1.grid()
ax1.set_title(r"$y_1 = \cos(2\pi x)$")
ax1.set_ylim(-2,2)
ax1.set_xlabel("x")
ax1.set_ylabel("y")

# # Build the third plot
ax2.plot(x, y2, 'g:')
ax2.grid()
ax2.set_title(r"$y_2 = y_0 + y_1$")
ax2.set_ylim(-2,2)
ax2.set_xlabel("x")
ax2.set_ylabel("y")

# # Build the last plot
ax3.plot(x, y3, 'k-')
ax3.grid()
ax3.set_title(r"$y_3 = y_0 - y_1$")
ax3.set_ylim(-2,2)
ax3.set_xlabel("x")
ax3.set_ylabel("y")

fig.tight_layout()
plt.show()
```

## Esempi con numpy e matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 3, 20)
y = np.linspace(0, 9, 20)
plt.plot(x, y)
plt.plot(x, y, "o")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(27446968)
x = np.linspace(0, 1, 20)
y = np.cos(x) + 0.3 * rng.random(20)
p = np.poly1d(np.polyfit(x, y, 3))
t = np.linspace(0, 1, 200)
plt.plot(x, y, "o", t, p(t), "-")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(27446968)
x = np.linspace(-1, 1, 2000)
y = np.cos(x) + 0.3 * rng.random(2000)
p = np.polynomial.Chebyshev.fit(x, y, 90)
plt.plot(x, y, "r.")
plt.plot(x, p(x), "k-", lw=3)
plt.show()
```
