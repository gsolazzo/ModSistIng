## Armonografo

Sono disegni prodotti da un [Armonografo](https://www.youtube.com/watch?v=fZCZZ5BbHWs)

Un armonografo  un dispositivo meccanico utilizzato per produrre figure geometriche complesse, spesso a forma di curve sinusoidali, attraverso l'interazione di movimenti armonici.

Questi dispositivi furono inventati nel XIX secolo come strumenti per visualizzare le propriet matematiche delle onde sonore e le loro interazioni. Oggi, sono principalmente usati a fini artistici e educativi.

Un armonografo di base  composto da una serie di componenti, tra cui un telaio, un pendolo o pi pendoli, e dispositivi di trascinamento come penne o matite. I pendoli oscillano in modo armonico, mentre i dispositivi di trascinamento applicano il movimento alle penne, creando cos le figure geometriche complesse su una superficie.

## La matematica dell'armonografo

Teoria delle Curve Parametriche: Le curve create dagli armonografi possono essere descritte matematicamente come curve parametriche, dove le coordinate della curva sono funzioni dei parametri che descrivono il movimento dei pendoli e dei dispositivi di trascinamento.

Ad esempio, per un armonografo con un pendolo e un dispositivo di trascinamento, le equazioni parametriche potrebbero essere:

\begin{equation} x(t) = A_1 \cos(\omega_x t + \phi_x) \end{equation}
\begin{equation} y(t) = B_1 \cos(\omega_y t + \phi_y) \end{equation}


Queste equazioni descrivono *le curve chiamate di Lissajous (o figure di Bowditch o curve del gioco di Alice)* e dipendono appunto da vari parametri:
1. le ampiezze A e B di tali moti;
2. le frequenze angolari: $_x, _y$
3. le fasi: $_x, _y$

## Simuliamo il funzionamento di un Armonografo con Python

Questo codice Python genera un grafico che rappresenta un insieme di punti $(X, Y)$ derivati da funzioni matematiche, e poi li visualizza.

```python
import numpy as np
import matplotlib.pyplot as plt

# NON MODIFICARE
n = 1000000
t = np.linspace(0, 100, 20000000)

# Definizione dei parametri
A = 1.5  # ampiezza x
B = 0.8  # ampiezza y
omega_x = 2.0  # frequenza angolare x
omega_y = 1.5  # frequenza angolare y
phi_x = np.pi / 2  # fase x (in radianti)
phi_y = np.pi / 3  # fase y (in radianti)

# Calcolo delle coordinate x e y
x = A * np.cos(omega_x * t + 2*phi_x)
y = B * np.cos(omega_y * t + 0.5*phi_y)

# Plot delle curve
plt.plot(x,y,'k',linewidth=.1)
plt.axis('off')
plt.show()
```

## Simuliamo il funzionamento di due Armonografi

Supponiamo ora di avere due Armonografi descritti dalle seguenti equazioni:

\begin{equation} x(t) = A_0 \sin(ft) e^{-d_0 t} + A_1 \sin(ft) e^{-d_1 t} \end{equation}
\begin{equation} y(t) = A_2 \sin(ft) e^{-d_2 t} + A_3 \sin(ft) e^{-d_3 t}\end{equation}

in cui:
* $A_0, A_1, A_2, A_3$ sono le ampiezze delle oscillazioni dei moti nelle due componenti
* $f_0, f_1, f_2, f_3$ sono le frequenze angolari  delle oscillazioni nelle due componenti
* $d_0, d_1, d_2, d_3$ sono sono i coefficienti di smorzamento nelle due componenti
* $t$  il tempo.



```python
# NON MODIFICARE
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)
```

*   la prima riga definisce il numero n di punti da generare per il grafico, impostato a 1.000.000. Questo significa che verranno calcolati un milione di punti $(X, Y)$ basati sulle equazioni fornite.
*   la seconda riga crea un array t di n punti logaritmicamente spaziati (uniformemente su una scala logaritmica) tra 10 e 500. L' array t conterr n valori distribuiti in modo uniforme su una scala logaritmica nell'intervallo da $10^1$ a $10^{2.69897}$. Questo  utile quando si desidera campionare l'asse del tempo o un altro intervallo in modo da coprire una vasta gamma di valori con una maggiore risoluzione in alcune parti dell'intervallo rispetto ad altre.

```python
# try changing these parameters
A = [  1, 1, 1.5, 1.5 ]
d = [ .004, .001, .002, .0015 ]
f = [   3, 1, 2, 2.5 ]

# generate XY value pairs
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)
```



*   A  un array contenente le ampiezze per le funzioni sinusoidali. Ogni elemento in A corrisponde all'ampiezza di una diversa sinusoide che contribuir al calcolo di x o y.
*   d  un array di coefficienti di smorzamento per le funzioni esponenziali. Questi valori influenzano la velocit con cui l'ampiezza delle sinusoidi decresce nel tempo.
*   f  un array con le frequenze per le funzioni sinusoidali. Ogni valore in f determina la frequenza di una delle sinusoidi utilizzate per calcolare x o y.



*   Calcola i valori di x come la somma di due sinusoidi smorzate. Ogni sinusoide  modulata in ampiezza da un'esponenziale negativa, che dipende dal tempo t, dalla frequenza f, dall'ampiezza A e dal coefficiente di smorzamento d.
*   Calcola i valori di y usando un approccio simile a quello usato per x, combinando due diverse sinusoidi smorzate.







Disegnamo i punti togliendo il riferimento degli assi e usando il colore nero

```python
# plot them!
plt.plot(x,y,'k',linewidth=.1)
plt.axis('off')
plt.show()
```
